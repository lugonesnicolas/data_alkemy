```python
import pandas as pd
```

```python
df_bibliotecas = pd.read_csv('datos/biblioteca/2022-05/biblioteca-06-05-2022/biblioteca-06-05-2022.csv',sep=',', header=0)
```

```python
df_bibliotecas = df_bibliotecas.rename({'Cod_Loc':'cod_localidad','IdProvincia':'id_provincia','IdDepartamento':'id_departamento','Categoría':'categoría','Provincia':'provincia','Localidad':'localidad','Nombre':'nombre','Domicilio':'domicilio',
                                               'CP':'código postal','Teléfono':'número de teléfono','Mail':'mail','Web':'web'},axis=1)
```

```python
df_bibliotecas.drop_duplicates()
```

```python
df_bibliotecas['año_inicio'] = df_bibliotecas['año_inicio'].fillna(0)
```

```python
df_bibliotecas['año_inicio'] = df_bibliotecas['año_inicio'].round(decimals=0).astype(int)
```

```python
df_bibliotecas.loc[df_bibliotecas.año_inicio==0,'año_inicio']='s/d'
```

```python
df_bibliotecas = df_bibliotecas.fillna('s/d')
```
