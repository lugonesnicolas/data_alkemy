```python
import pandas as pd
```

```python
df_cines = pd.read_csv('datos/cine/2022-05/cine-06-05-2022/cine-06-05-2022.csv',sep=',',header=0)  #Importamos nuestro data frame
```

```python
df_cines_count = df_cines.groupby(['provincia'],as_index=None).count()[['provincia','espacio_INCAA']]
```

```python
df_cines_coun_incaa = df_cines_count[['espacio_INCAA']]
```

```python
df_cines_coun_incaa
```

```python
df_cines_concat_prov = df_cines.groupby(['provincia'],as_index=None).count()[['provincia']]
```

```python
df_cines_count_pantallas = df_cines.groupby(['provincia'],as_index=None)['Pantallas'].apply(lambda x : x.astype('int64').sum())
df_cines_count_pantallas = df_cines_count_pantallas[['Pantallas']]
```

```python
df_cines_count_butacas = df_cines.groupby(['provincia'],as_index=None)['Butacas'].apply(lambda x : x.astype('int64').sum())
df_cines_count_butacas = df_cines_count_butacas[['Butacas']]
```

```python
df_cines_concat_prov
```

```python
df_cines_count_pantallas
```

```python
df_cines_count_butacas
```

```python
df_concat_count_cines = pd.concat([df_cines_concat_prov, df_cines_count_pantallas], axis=1)
```

```python
df_concat_count_cines = pd.concat([df_concat_count_cines, df_cines_count_butacas], axis=1)
```

```python
df_concat_count_cines = pd.concat([df_concat_count_cines, df_cines_coun_incaa], axis=1)
```
