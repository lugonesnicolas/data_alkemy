```python
import pandas as pd
```

```python
df_museos = pd.read_csv('datos/museos_datosabiertos/2022-05/museos_datosabiertos-08-05-2022.csv',sep=',',header=0)
```

```python
df_museos
```

```python
df_museos['año_inauguracion'] = df_museos['año_inauguracion'].fillna(0) #convertimos los Nan en 0 para poder parlos a enteros
```

```python
df_museos['año_inauguracion'] = df_museos['año_inauguracion'].round(decimals=0).astype(int) #pasamos los datos de la columna a enteros
```

```python
df_museos.loc[df_museos.año_inauguracion==0,'año_inauguracion']='s/d' #los datos 0 los ponemos como sin dato
```

```python
df_museos = df_museos.fillna('s/d')
```

```python
df_museos = df_museos.rename({'Cod_Loc':'cod_localidad','IdProvincia':'id_provincia','IdDepartamento':'id_departamento','categoria':'categoria','Provincia':'provincia','Localidad':'localidad','Nombre':'nombre','direccion':'domicilio',
                                               'CP':'codigo postal','telefono':'numero de telefono','Mail':'mail','Web':'web'},axis=1)
```
