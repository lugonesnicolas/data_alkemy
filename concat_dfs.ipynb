```python
import pandas as pd
```

```python
df_bibliotecas = pd.read_csv('datos/biblioteca/2022-05/biblioteca-06-05-2022/biblioteca-06-05-2022.csv',sep=',', header=0)
df_cines = pd.read_csv('datos/cine/2022-05/cine-06-05-2022/cine-06-05-2022.csv',sep=',',header=0)  #Importamos nuestro data frame
df_museos = pd.read_csv('datos/museos_datosabiertos/2022-05/museos_datosabiertos-06-05-2022/museos_datosabiertos-06-05-2022.csv',sep=',',header=0)
```

```python
df_bibliotecas
```

```python
df_bibliotecas_normalized = df_bibliotecas[['cod_localidad','id_provincia','id_departamento','categoria','provincia','localidad','nombre','domicilio','codigo postal','numero de telefono','mail','web']]
```

```python
df_cines_normalized = df_cines[['cod_localidad','id_provincia','id_departamento','categoria','provincia','localidad','nombre','domicilio','codigo postal','numero de teléfono','mail','web']]
```

```python
df_museos_normalized = df_museos[['cod_localidad','id_provincia','id_departamento','categoria','provincia','localidad','nombre','domicilio','codigo postal','numero de teléfono','mail','web']]
```

```python
dfs_concat = pd.concat([df_bibliotecas_normalized, df_cines_normalized, df_museos_normalized], axis=0)
```

```python
dfs_concat
```

```python
dfs_concat.to_csv('datos_norm/data_norm.csv', index=None)
```

# Datos por categoria y fuente

```python
dfs_concat.groupby('categoría', as_index=False).count()[['categoría', 'provincia']]
```

```python
df_fuentes_categoria = dfs_concat.groupby('categoría', as_index=False).count()[['categoría', 'provincia']]
fuentes =['bibliotecas', 'museos', 'cines']
fuentes_add = df_fuentes_categoria
fuentes_add['fuentes'] = fuentes
```

```python
fuentes_add_df = pd.DataFrame(fuentes_add)
```

```python
fuentes_add_df
```

```python
fuentes_add_df.to_csv('datos_norm/registros_categoria_fuentes.csv', index=None)
```
